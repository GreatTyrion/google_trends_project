```python
def transfer_time(trend_data):
    timestamp = int(trend_data['time'])
    dt_object = datetime.fromtimestamp(timestamp)
    date_time = dt_object.strftime("%Y-%m-%d")
    return date_time
```

```python
for item in data['default']['timelineData']:
    print(transfer_time(item), item['value'][0])
```
